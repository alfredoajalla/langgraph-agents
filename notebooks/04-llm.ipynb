```python
import os 
from dotenv import load_dotenv

load_dotenv()


```

```python
from langchain_openai import ChatOpenAI

gpt4_llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key=os.environ.get("OPENAI_API_KEY"),
)

response = gpt4_llm.invoke("Hello from Bs. As.")
response.pretty_print()
```

```python
from langchain_anthropic import ChatAnthropic

claude_llm = ChatAnthropic(
    model="claude-3-5-sonnet-20240620",
    temperature=0,
    anthropic_api_key=os.environ.get("ANTROPIC_API_KEY"),
)

response = claude_llm.invoke("Hello from Bs. As.")
response.pretty_print()
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage

msg1 = HumanMessage(content="Hello, how are you?")
msg2 = SystemMessage(content="You are a helpful assistant.")
msg3 = AIMessage(content="Hello! I'm doing well, thank you for asking.")
msg4 = HumanMessage(content="Can you tell me a joke?")
history = [msg1, msg2, msg3, msg4]
response = gpt4_llm.invoke(history)
response.pretty_print()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key=os.environ.get("OPENAI_API_KEY"),
)
response = llm.invoke(history)
response.pretty_print()
```
