```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.chat_models import init_chat_model
import os

load_dotenv()

llm = init_chat_model(
    model="gpt-4o",
    temperature=0,
    openai_api_key=os.environ.get("OPENAI_API_KEY"),
    streaming=False,
)
```

```python
file_search_tool = {
    "type": "file_search",
    "vector_store_ids": ["vs_6a61864d3738819194cfe35494ffb3bb"],    
}

llm_with_tools = llm.bind_tools([file_search_tool])
```

```python
response = llm_with_tools.invoke("Cuales son las normas de uso del float en la construccion?")
response.pretty_print()
```
