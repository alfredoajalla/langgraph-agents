```python
from langchain_core.messages import AIMessage 

ai_message = AIMessage(content="Hello, how can I help you? Message from AI")

ai_message.text
```

```python
from langchain_core.messages import HumanMessage 

human_message = HumanMessage(content="Hello, I need information about your services. Message from Human")

human_message.text
```

```python
history = [ai_message, human_message]

for m in history:
    m.pretty_print()

```
